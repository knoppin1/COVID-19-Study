
Tokenize the 'SpaceX.txt' corpus

```python
# Load text using spacy
en = spacy.load('en_core_web_sm')
corpus = en(text)
print(corpus)
```

```python
# Tokenize the 'SpaceX.txt' corpus
tokens = [token.text for token in corpus]
print(tokens)
```

```python
# Remove stop words and punctuation using spacy
filtered_list = [token.text for token in corpus if token.is_stop == False and token.is_punct == False] 
print(filtered_list)
```

Determine maximum frequency (as word_frequencies[word] = (word_frequencies[word]/maximum_frequency)

```python
# Determine word frequencies
from collections import Counter
word_frequencies = Counter(filtered_list)
word_frequencies.most_common(5)
```

```python
# Determine maximum frequency of a word in the corpus.
maximum_frequency = Counter(filtered_list).most_common(1)[0][1]
maximum_frequency
```

```python
# Normalize other words based on the maximum frequency.
for word in word_frequencies.keys():
    word_frequencies[word] = (word_frequencies[word]/maximum_frequency)
word_frequencies.most_common(5)
```

```python
# Alternative of above normalization using NLTK (not used - for reference only)
import nltk
word_frequencies = nltk.FreqDist(filtered_list)
maximum_frequency = word_frequencies[word_frequencies.max()]
for word in word_frequencies:
    word_frequencies[word] = word_frequencies[word]/maximum_frequency
word_frequencies.most_common(5)
```

Tokenize the sentences. Generate the sentence_scores. Score every sentence
based on number of words.

```python
print('Number of sentences:', len(list(corpus.sents)))
```

```python
sentence_score = {}
for sentence in corpus.sents:
    for word in sentence:
        if word.text in word_frequencies.keys():
            if sentence in sentence_score.keys():
                sentence_score[sentence] += 1 # Increment word count
            else:
                sentence_score[sentence] = 1 # Initialize word count
print(sentence_score)                
```

```python
# Alternative - Load a dataframe with each sentence and scores.  (Not used - for reference only)
import pandas as pd
df = pd.DataFrame(columns=['Sentence','Raw_Score','Clean_Score'])
for sentence in corpus.sents:

    # Tokenize sentence and determine a raw score for all words in the sentence only removing punctuation.
    words = [token.text for token in sentence if token.is_punct != True]
    raw_score = len(words)

    # Tokenize sentence and determine a clean score for all words in the sentence after removing stop words and punctuation.
    words = [token.text for token in sentence if token.is_stop != True and token.is_punct != True]
    clean_score = len(words)

    # Add row to dataframe
    df = df.append({'Sentence': ''.join(token.string for token in sentence), 'Raw_Score': raw_score, 'Clean_Score': clean_score}, ignore_index=True)
df
```

Import nlargest from heapq and provide 'summarized_sentences'.

```python
# Get a summary of sentences based on the ten highest scores
from heapq import nlargest
summarized_sentences = nlargest(10, sentence_score, key=sentence_score.get)
for s in summarized_sentences:
    print(s)
```

Convert sentences from spacy to strings and join all sentences

```python
# Convert spacy sentences to strings
sentences = []
for sentence in summarized_sentences:
    sentences.append(''.join(token.string for token in sentence))
sentences    

# Join sentences
joined_sentences = ''.join(sentences)
joined_sentences
```

Provide a summary once you have converted spacy outputs to strings.

```python
summarized_sentences_string_list = [word.text for word in summarized_sentences]
summary = ' '.join(summarized_sentences_string_list)
print(summary)
```

Determine the length of the summary.

```python
print("Summary Length in Words:", len([token.text for token in en(summary)]))
```
